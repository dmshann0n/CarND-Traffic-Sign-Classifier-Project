### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance. 

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.
### Feel free to use as many code cells as needed.

import cv2
import random

CROP_SIZE = 2

def rescale(img):
    # if already at 32x32, we can probably do some cropping!
    if img.shape[0] == 32 and img.shape[1] == 32:
        img = img[CROP_SIZE:img.shape[0] - CROP_SIZE, CROP_SIZE:img.shape[1] - CROP_SIZE]

    return cv2.resize(img, (32, 32), interpolation=cv2.INTER_CUBIC)

def contrast_img(img, clip=3, grid_size=2):
    # from research, found here: https://stackoverflow.com/questions/19363293
    clahe = cv2.createCLAHE(clipLimit=clip, tileGridSize=(grid_size,grid_size))
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)  # convert from BGR to LAB color space
    l, a, b = cv2.split(lab)  # split on 3 different channels

    l2 = clahe.apply(l)  # apply CLAHE to the L-channel

    lab = cv2.merge((l2,a,b))  # merge channels
    return cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)  # convert from LAB to BGR

def jitter_img(img):
        
    rows, cols, _ = img.shape
    
    # follow the Pierre Sermanet and Yann LeCun process
    # "Samples are randomly perturbed in position ([-2,2] pixels), in scale ([.9,1.1] ratio) 
    # and rotation ([-15,+15] degrees)."
    
    new_img = img.copy()
    
    # translate by -2 to 2 pixels
    x_translate = random.randrange(-2, 2)
    y_translate = random.randrange(-2, 2)
    
    M = np.float32([[1, 0, x_translate], [0, 1, y_translate]])
    new_img = cv2.warpAffine(new_img, M, (cols, rows))   
    
    # scale by .9 to 1.1 (process_img will resize back to 32x32 later)
    x_scale = random.uniform(.9, 1.1)
    y_scale = random.uniform(.9, 1.1)
    
    new_img = cv2.resize(new_img, (int(rows * x_scale), int(cols * y_scale)), interpolation=cv2.INTER_CUBIC)  
    
    # rotating randomly -15 to 15 degrees
    rotation = random.uniform(-15, 15)
    M = cv2.getRotationMatrix2D((cols/2, rows/2), rotation, 1)    
    new_img = cv2.warpAffine(new_img, M, (cols, rows))
    
    # also suggested to jitter sharpness and contrast
    blur = cv2.GaussianBlur(new_img, (0,0), 3)
    new_img = cv2.addWeighted(new_img, 2, blur, random.uniform(-1,1), 0)
    
    new_img = contrast_img(new_img, random.uniform(0, 40), random.randrange(1, 8))
    
    return new_img
    
def process_img(original):
    img = original.copy()
    filters = [contrast_img, rescale]
    for f in filters:
        img = f(img)
    
    return img

def process_batch(img_batch):
    new_img = []
    for img in img_batch:
        new_img.append(process_img(img))
        
    return new_img

plt.figure(figsize=(4,2))
img = X_train[y_train == 1][0]
processed = process_img(img)

print('Original dimensions: {}'.format(img.shape))
print('New dimensions: {}'.format(processed.shape))

plt.subplot(1, 3, 1)
plt.title('original')
plt.imshow(img)

plt.subplot(1, 3, 2)
plt.title('pre-processed')
plt.imshow(processed)

plt.subplot(1, 3, 3)
plt.title('jitter')
plt.imshow(process_img(jitter_img(img)))

plt.show()
````

````python
X_train_processed = process_batch(X_train)
print('Modified training set. Length: {}'.format(len(X_train_processed)))

X_valid_processed = process_batch(X_valid)
print('Modified validation set. Length: {}'.format(len(X_valid_processed)))        
````

````python
from sklearn.utils import shuffle

def get_training_data():
    
    X_train_res = X_train_processed.copy()
    y_train_res = []
    
    PERMS = 10
    for i, img in enumerate(X_train):
        label = y_train[i]
        for n in range(PERMS):
            X_train_res.append(process_img(jitter_img(img)))
            y_train_res.append(label)

    y_train_res = np.append(y_train, y_train_res)
        
    print('Jittered training set. Length: {}'.format(len(X_train_res)))
    
    return shuffle(X_train_res, y_train_res)
````

### Model Architecture

````python
import tensorflow as tf
from tensorflow.contrib.layers import flatten

class LeNetWrapper:
    def __init__(self, logits, *weights):
        self.logits = logits
        self.weights = weights

def LeNet(x):    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # SOLUTION: Layer 1: Convolutional. Input = 32x32x3. Output = 28x28x6.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 3, 6), mean = mu, stddev = sigma), name='conv1_W')
    conv1_b = tf.Variable(tf.zeros(6))
    conv1   = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID') + conv1_b

    # SOLUTION: Activation.
    conv1 = tf.nn.relu(conv1)

    # SOLUTION: Pooling. Input = 28x28x6. Output = 14x14x6.
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # SOLUTION: Layer 2: Convolutional. Output = 10x10x16.
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 6, 16), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(16))
    conv2   = tf.nn.conv2d(conv1, conv2_W, strides=[1, 1, 1, 1], padding='VALID') + conv2_b
    
    # SOLUTION: Activation.
    conv2 = tf.nn.relu(conv2)

    # SOLUTION: Pooling. Input = 10x10x16. Output = 5x5x16.
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    # SOLUTION: Flatten. Input = 5x5x16. Output = 400.
    fc0   = flatten(conv2)
       
    # SOLUTION: Layer 3: Fully Connected. Input = 400. Output = 120.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(400, 120), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(120))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    
    # SOLUTION: Activation.
    fc1    = tf.nn.relu(fc1)
    
    fc1   = tf.nn.dropout(fc1, dropout_rate)
    
    # SOLUTION: Layer 4: Fully Connected. Input = 120. Output = 84.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(120, 84), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(84))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    # SOLUTION: Activation.
    fc2    = tf.nn.relu(fc2)

    # SOLUTION: Layer 5: Fully Connected. Input = 84. Output = 10.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(84, n_classes), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(n_classes))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return LeNetWrapper(logits, conv1_W, conv2_W, fc2_W, fc3_W)
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.

rate = 0.0005
EPOCHS = 15
BATCH_SIZE = 128
dropout = 0.5

x = tf.placeholder(tf.float32, (None, 32 , 32, 3))
y = tf.placeholder(tf.int32, (None))
dropout_rate = tf.placeholder(tf.float32)
one_hot_y = tf.one_hot(y, n_classes)

network = LeNet(x)
logits = network.logits
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)

correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, dropout_rate: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples

X_train_all, y_train_all = get_training_data()

saver = tf.train.Saver()
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        print("Starting EPOCH {}".format(i+1))
        X_train_all, y_train_all = shuffle(X_train_all, y_train_all)
        
        num_examples = len(X_train_all)
        num_batches = num_examples // BATCH_SIZE
        validation_at = num_batches // 4 * BATCH_SIZE
        
        
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train_all[offset:end], y_train_all[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, dropout_rate: dropout})
            
            if (offset > 0 and offset % validation_at == 0):
                validation_accuracy = evaluate(X_valid_processed, y_valid)
                print("Validation Accuracy @ {} inputs = {:.3f}".format(end, validation_accuracy))
                
        print()

    validation_accuracy = evaluate(X_valid_processed, y_valid)
    print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        
    saver.save(sess, './traffic_sign')
    print("Model saved")
````

## Step 4 (Optional): Visualize the Neural Network's State with Test Images

 This Section is not required to complete but acts as an additional excersise for understaning the output of a neural network's weights.
